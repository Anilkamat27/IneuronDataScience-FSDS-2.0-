```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
X,y = make_moons(n_samples=250,noise=0.05,random_state=42)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=100)
plt.show()
```

```python
model = Sequential()
model.add(Dense(10,activation='sigmoid',input_dim=2))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer ='adam',metrics=['accuracy'])
```

```python
model.get_weights()[0]
```

```python
old_weights = model.get_weights()[0]
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
model.fit(X_train,y_train,epochs = 300)
```

```python
new_weights= model.get_weights()[0]
```

```python
new_weights
```

```python
old_weights
```

```python
percentage_change = abs(100*(old_weights - new_weights)/old_weights)
```

It is clearly visible in Deep Neural network , change in weights is very very small , facing problem of vanishing gradient

reason is deep ANN and sigmoid Function

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train,y_train.astype('int'),clf=model,legend=2)
```

 to visualize the decision boundaries of a classifier.

so it is clearly visble that there is no decison boundaries in graph , means , my model is unsuccesfull in learning due to vanishing gradient problem


So how to overcome this type of problem

First solution : reduce the model time complexity

```python
model2 = Sequential()
model2.add(Dense(10,activation='sigmoid',input_dim=2))
model2.add(Dense(10,activation='sigmoid'))
model2.add(Dense(10,activation='sigmoid'))
model2.add(Dense(10,activation='sigmoid'))
model2.add(Dense(10,activation='sigmoid'))
model2.add(Dense(1,activation='sigmoid'))
```

```python
model2.compile(loss='binary_crossentropy',optimizer ='adam',metrics=['accuracy'])
model2.fit(X_train,y_train,epochs = 100)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train,y_train.astype('int'),clf=model2,legend=2)
```

Solution 2 : Using Relu Activation function

```python
model3 = Sequential()
model3.add(Dense(10,activation='relu',input_dim=2))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(1,activation='sigmoid'))
```

```python
model3.compile(loss='binary_crossentropy',optimizer ='adam',metrics=['accuracy'])
model3.fit(X_train,y_train,epochs = 100)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train,y_train.astype('int'),clf=model3,legend=2)
```

Solution 3rd : weight Initialization

```python
model4 = Sequential()
model4.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal'))
model4.add(Dense(10,activation='relu' , kernel_initializer='he_normal'))
model4.add(Dense(10,activation='relu', kernel_initializer='he_normal'))
model4.add(Dense(10,activation='relu' , kernel_initializer='he_normal'))
model4.add(Dense(10,activation='relu', kernel_initializer='he_normal'))
model4.add(Dense(1,activation='sigmoid'))
```
