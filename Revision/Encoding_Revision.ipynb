**Encoding categorical data**

categorical data is divided into two parts

NOMINAL : JAHA PE categories me koi relationship nahi hota nahi koi order hota hai , like states

ORDINAL : jaha pe catgories me relationship hota hai , order hota hai , like result of students , like divisons
, remarks like excellent , good , bad

Two most important types of encoding is

Ordinal encoding : works on ordinal data

one hot encoding : works on nominal data

one more type of encoding is label encoding similar to ordinal encoding , there is only small difference , agar output variable hi categorical type ka hai toh waha ordinal nahi laga sakte , humme waha label encoding use karni padti hai , for this there is special class available called label encoder in scikit-learn
